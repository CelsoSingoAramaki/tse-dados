```mysql
SELECT 
ANO_ELEICAO,NUM_TURNO,DESCRICAO_ELEICAO,SIGLA_UF,SIGLA_UE,CODIGO_MUNICIPIO,NOME_MUNICIPIO,CODIGO_CARGO,DESCRICAO_CARGO,NUM_VOTAVEL from votacao_secao_2016
where (NUM_VOTAVEL="95" OR NUM_VOTAVEL="96" OR NUM_VOTAVEL="97") AND 
group by ANO_ELEICAO,NUM_TURNO,DESCRICAO_ELEICAO,SIGLA_UF,SIGLA_UE,CODIGO_MUNICIPIO,NOME_MUNICIPIO,CODIGO_CARGO,DESCRICAO_CARGO,NUM_VOTAVEL
order by NUM_VOTAVEL;
```
