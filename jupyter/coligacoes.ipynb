```python
import pandas as pd

df = pd.read_csv('legendas.csv', sep=';')
df.head()
```

```python
parties_keys = ['ano_eleicao', 'num_turno', 'sigla_partido', 'nome_partido']
parties_jobs_keys = ['ano_eleicao', 'num_turno', 'sigla_partido', 'nome_partido', 'codigo_cargo', 'descricao_cargo']
parties_jobs = df.groupby(parties_jobs_keys).agg({
    "composicao_coligacao": lambda x: len(x.str.split("/")) - 1
}).reset_index()
parties_jobs = parties_jobs.set_index(parties_keys)
```

```python
#PRESIDENTE
agg1 = parties_jobs[parties_jobs.codigo_cargo == 1].rename(columns={'composicao_coligacao': 'presidente'})
agg1 = agg1[['presidente']]
agg1
```

```python
#GOVERNADOR
agg3 = parties_jobs[parties_jobs.codigo_cargo == 3].rename(columns={'composicao_coligacao': 'governador'})
agg3 = agg3[['governador']]
agg3
```

```python
#SENADOR
agg5 = parties_jobs[parties_jobs.codigo_cargo == 5].rename(columns={'composicao_coligacao': 'senador'})
agg5 = agg5[['senador']]
agg5
```

```python
#DEP FEDERAL
agg6 = parties_jobs[parties_jobs.codigo_cargo == 6].rename(columns={'composicao_coligacao': 'deputado_federal'})
agg6 = agg6[['deputado_federal']]
agg6
```

```python
#DEP ESTADUAL
agg7 = parties_jobs[parties_jobs.codigo_cargo == 7].rename(columns={'composicao_coligacao': 'deputado_estadual'})
agg7 = agg7[['deputado_estadual']]
agg7
```

```python
#DEP DISTRITAL
agg8 = parties_jobs[parties_jobs.codigo_cargo == 8].rename(columns={'composicao_coligacao': 'deputado_distrital'})
agg8 = agg8[['deputado_distrital']]
agg8
```

```python
#PREFEITO
agg11 = parties_jobs[parties_jobs.codigo_cargo == 11].rename(columns={'composicao_coligacao': 'prefeito'})
agg11 = agg11[['prefeito']]
agg11
```

```python
#VEREADOR
agg13 = parties_jobs[parties_jobs.codigo_cargo == 13].rename(columns={'composicao_coligacao': 'vereador'})
agg13 = agg13[['vereador']]
agg13
```

```python
final = agg1.join(agg3, how='outer')\
            .join(agg5, how='outer')\
            .join(agg6, how='outer')\
            .join(agg7, how='outer')\
            .join(agg8, how='outer')\
            .join(agg11, how='outer')\
            .join(agg13, how='outer')
final.reset_index()
final['governador'] = final.governador.fillna(0).astype(object)
final['senador'] = final.senador.fillna(0).astype(object)
final['deputado_federal'] = final.deputado_federal.fillna(0).astype(object)
final['deputado_estadual'] = final.deputado_estadual.fillna(0).astype(object)
final['deputado_distrital'] = final.deputado_distrital.fillna(0).astype(object)
final['prefeito'] = final.prefeito.fillna(0).astype(object)
final['vereador'] = final.vereador.fillna(0).astype(object)
final = final.reset_index()
final
```

```python
final.to_csv('coligacoes.csv', sep=';', index=False, header=True)
```
