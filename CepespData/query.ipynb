```python
from web.cepesp.athena.builders import AthenaBuilder
```

```python
def query(**options):
    builder = AthenaBuilder(**options)
    print(builder.build())
```

```python
query(table='votos', years=[2014], job=7, reg=4, brancos=True, uf_filter='CE')
```

```python
query(table='candidatos', years=[2014], job=1, only_elected=True)
```

```python
query(table='candidatos', years=[2014], job=7, filters={
    'NOME_CANDIDATO': 'FULANO DE TAL'
})
```
