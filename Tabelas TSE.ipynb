# Relacionando tabela de votacao_secao com a tabela detalhe_candidatos


### v = VOTACAO_SECAO
### c = CANDIDATO

* v.ANO_ELEICAO=c.ANO_ELEICAO
* v.SIGLA_UE=c.SIGLA_UE 
* v.CODIGO_CARGO=c.CODIGO_CARGO 
* v.NUM_VOTAVEL=c.NUMERO_CANDIDATO 
* v.NUM_TURNO=c.NUM_TURNO;

#### Sql votacao_secao com detalhe candidato 2014_2016, agregando por macro regiao

``` mysql
select * from 
(
    select d.CODIGO_CARGO as CODIGO_CARGO_V,
    d.DESCRICAO_CARGO as DESCRICAO_CARGO_V,
    d.NUM_VOTAVEL,
    d.NUM_TURNO as NUM_TURNO_V,
    d.DESCRICAO_ELEICAO as DESCRICAO_ELEICAO_V,
    d.ANO_ELEICAO as ANO_ELEICAO_V,
    sum(cast(QTDE_VOTOS as int)) as TOTAL_VOTOS,
    d.macro_, d.macro_des, d.SIGLA_UE as SIGLA_UE_V from (
        select 
        HORA_GERACAO,
        ANO_ELEICAO,
        NUM_TURNO,
        DESCRICAO_ELEICAO,
        CODIGO_MUNICIPIO,
        CODIGO_CARGO,
        DESCRICAO_CARGO,
        NUM_VOTAVEL,
        QTDE_VOTOS,
        SIGLA_UE,
        macro_,macro_des from (
            select * from votacao_secao
            ) as a join aux_municipio as m on a.CODIGO_MUNICIPIO=m.cod
        ) as d group by d.CODIGO_CARGO,
        d.DESCRICAO_CARGO,
        d.NUM_VOTAVEL,
        d.NUM_TURNO,
        d.DESCRICAO_ELEICAO,
        d.ANO_ELEICAO,
        d.SIGLA_UE,
        d.macro_,
        d.macro_des
) as v
join detalhe_candidato_2014_2016 as c 
on v.ANO_ELEICAO_V=c.ANO_ELEICAO AND 
v.SIGLA_UE_V=c.SIGLA_UE AND 
v.CODIGO_CARGO_V=c.CODIGO_CARGO AND 
v.NUM_VOTAVEL=c.NUMERO_CANDIDATO AND 
v.NUM_TURNO_V=c.NUM_TURNO;
```
