```python
import pandas as pd
from glob import glob

vot = None
for f in glob('etl/output/final/votos/uf/2002/6/**/*.gz'):
    df = pd.read_csv(f, sep=';', dtype=str)
    vot = df if vot is None else vot.append(df, ignore_index=True)
vot.head()
```

```python
cand = pd.read_csv('etl/output/final/candidatos/2002/6/candidato_2002.gz', sep=';', dtype=str)
```

```python
len(vot), len(cand)
```

```python
elected = cand[(cand.COD_SIT_TOT_TURNO == '1') | (cand.COD_SIT_TOT_TURNO == '2') | (cand.COD_SIT_TOT_TURNO == '3')]
```

```python
len(elected)
```

```python
vot_cand = vot.set_index(['ID_CANDIDATO']).merge(cand.set_index(['ID_CANDIDATO']), how='left', left_index=True, right_index=True)
```

```python
vot_cand_elected = vot_cand[(vot_cand.COD_SIT_TOT_TURNO == '1') | (vot_cand.COD_SIT_TOT_TURNO == '2') | (vot_cand.COD_SIT_TOT_TURNO == '3')]
```
