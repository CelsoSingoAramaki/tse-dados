````python
import pandas as pd

mun = pd.read_csv("data/aux_municipio.csv", delimiter=";",encoding="latin1")

````

````python
mun_tse = pd.read_csv("data/aux_mun_tse_code.csv", delimiter=";",encoding="latin1")

mun_tse["IBGEcod"] = mun_tse["IBGEcod"].real.astype(int)
mun_tse.groupby("Ano").count()
#mun_tse.set_index("IBGEcod").join(mun.set_index("mun_cod"))
````

### 

````python
mun_ano = mun_tse.groupby(["SiglaUF","IBGEcod","TSEcod"],as_index=False).max()
mun_ano = mun_ano.drop(["Ano"],1)
````

````python
mun_ano.head()
````

### Juntando as duas tabelas

````python
mun_ano_tse_cod = mun_ano.set_index("IBGEcod").join(mun.set_index("mun_cod"))
mun_ano_tse_cod.head()
````

````python
mun_ano_tse_cod = mun_ano_tse_cod.sort_values('TSEcod')
````

````python
mun_ano_tse_cod.head()
````

### Teste agregações micro e meso

````python
mun_ano_tse_cod.groupby(["meso","micro"])["TSEcod"].apply(list).head()
````

### Adicionando ao código TSE zeros a esquerda até ter 5 digitos

````python
mun_ano_tse_cod["TSEcod"] = mun_ano_tse_cod["TSEcod"].apply(lambda x: '{0:0>5}'.format(x))
mun_ano_tse_cod.head()
````

### Inserir dados de macro



````python
a = mun_ano_tse_cod.groupby(["CODIGO_MACRO","NOME_MACRO","SIGLA_LEGENDA","NOME_LEGENDA"],as_index=False)["QTDE_VOTOS"].sum()
a.sort_values('uf_sigla').head()
````

````python
macro_uf = pd.read_csv("data/macro_uf.csv")
macro_uf.set_index("sigla_UF")
````

### Adicionando colunas macro na tabela de municipio

````python
mun_ano_tse_cod = mun_ano_tse_cod.set_index("SiglaUF").join(macro_uf.set_index("sigla_UF"))

````
