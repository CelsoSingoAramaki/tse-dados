````python
import pandas as pd

data = pd.read_csv("data/local_votacao.csv.gz",compression="gzip", error_bad_lines=False, delimiter=";")
data.head()
````

````python
data = data.drop(["DT_EXTRACAO","CD_PROCESSO_ELEITORAL","QT_APTOS","CD_PLEITO","ST_AGREGADA","ST_ESPECIAL","ST_SECAO_USA_LOCVOT_TEMP"],1)
data.rename(columns={'SG_UF': 'UF', 'CD_LOCALIDADE_TSE': 'CODIGO_MUNICIPIO',
                     'NM_LOCALIDADE':'NOME_MUNICIPIO',
                     'NR_ZONA':'NUM_ZONA',
                     'NR_SECAO':'NUM_SECAO',
                     'NM_LOCVOT':'NOME_LOCAL_VOTACAO',
                     'DS_ENDERECO':'ENDERECO_LOCAL_VOTACAO',
                     'DS_BAIRRO':'BAIRRO_LOCAL_VOTACAO',
                     'NR_CEP': 'CEP_LOCAL_VOTACAO',
                     'NR_LOCVOT': 'NUMERO_LOCAL_VOTACAO'
                    }, inplace=True)
````

````python
import csv
data.to_csv("data/aux_local_votacao.csv",sep=';', encoding='utf-8',index=False,header=False, quoting=csv.QUOTE_ALL)
````
